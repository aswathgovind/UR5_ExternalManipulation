```python
from pynput.keyboard import Key, Listener
import rtde_control
import rtde_io
import rtde_receive

global robot
global robo_io
global rtde_r


def __del__(): 
        print('Destructor called, Employee deleted.') 
        if(robot.isConnected()):
            robot.disconnect()
            rtde_io.disconnect()
        if(rtde_r.isConnected()):
            rtde_r.disconnect()
        
        print("rtde.control, rtde_io,rtde_receive",robot.isConnected(),rtde_r.isConnected())
     
def on_press(key):
    if(key == Key.left):
        print("left arrow")
        move_robo([-0.1,0,0,0,0,0])
    elif(key == Key.right):
        print("right arrow")
        move_robo([0.1,0,0,0,0,0])
    elif(key == Key.down):
        print("down arrow")
        move_robo([0,-0.1,0,0,0,0])
    elif(key == Key.up):
        print("up arrow")
        move_robo([0,0.1,0,0,0,0])
    elif(key == Key.page_up):
        print("page_up")
        move_robo([0,0,0.1,0,0,0])
    elif(key == Key.page_down):
        print("page_down")
        move_robo([0,0,-0.1,0,0,0])
    elif(key == Key.ctrl_l):
        print("ctrl_l")
        rotate_robo([0,0,0,1,0,0])
    elif(key == Key.ctrl_r):
        print("ctrl_r")
        rotate_robo([0,0,0,-1,0,0])
    elif(key == Key.shift_l):
        print("shift_l")
        rotate_robo([0,0,0,0,-1,0])
    elif(key == Key.shift_r):
        print("shift_r")
        rotate_robo([0,0,0,0,1,0])
    elif(key == Key.alt_l):
        print("alt_l")
        rotate_robo([0,0,0,0,0,-1])
    elif(key == Key.alt_r):
        print("alt_r")
        rotate_robo([0,0,0,0,0,1])
   
def on_release(key):
    robot.speedStop()
    
    if key == Key.esc:
        return False

def move_robo(pose_wrt_tool):
    robo_actual_pose = rtde_r.getActualTCPPose()
    rel_robo_actual_pose = [0,0,0,robo_actual_pose[3],robo_actual_pose[4],robo_actual_pose[5]]
    speed_vect = robot.poseTrans(rel_robo_actual_pose,pose_wrt_tool)
    t_speed_vect = [speed_vect[0],speed_vect[1],speed_vect[2],0,0,0]
    #pose_wrt_base = robot.poseTrans(robot.getForwardKinematics(), pose_wrt_tool)
    robot.speedL(t_speed_vect,1.0,1.0)
    print(t_speed_vect)
    
def rotate_robo(pose_wrt_tool):
    robo_actual_pose = rtde_r.getActualTCPPose()
    rel_robo_actual_pose = [robo_actual_pose[0],robo_actual_pose[1],robo_actual_pose[2],0,0,0]
    speed_vect = robot.poseTrans(rel_robo_actual_pose,pose_wrt_tool)
    t_speed_vect = [0,0,0,speed_vect[3],speed_vect[4],speed_vect[5]]
    #pose_wrt_base = robot.poseTrans(robot.getForwardKinematics(), pose_wrt_tool)
    robot.speedL(t_speed_vect,1.0,1.0)
    print(t_speed_vect)    
    
robot = rtde_control.RTDEControlInterface("127.0.0.1")
robo_io = rtde_io.RTDEIOInterface("127.0.0.1")
rtde_r = rtde_receive.RTDEReceiveInterface("127.0.0.1")
print("rtde.control, rtde_io,rtde_receive",robot.isConnected(),rtde_r.isConnected())


with Listener(
        on_press=on_press,
        on_release=on_release) as listener:
    listener.join()
```
